```python
# IMPORTS
import pandas as pd
import numpy as np
#import matplotlib as mpl
```

<h1>PREPROCESS ALL ARC COUNTS TABLE</h1>

```python
# READ-IN ARC COUNT DATAFRAME
filename = 'data/MEBL3/MEBL3_All Arc Counts-data-as-joinbyfield-2024-07-01 10_16_54.csv'
arc_count_df = pd.read_csv(filename, parse_dates=['Time'])    
print(f"{filename} : file read into a pandas dataframe.")
```

```python
# DROP PRESSURE COLUMNS (STORED SEPARATELY)
arc_count_df2 = arc_count_df.drop(arc_count_df.columns[1:3], axis=1)
print(f"{len(arc_count_df.columns)} Columns")
arc_count_df2
```

```python
# USE REGEX TO REPLACE QUERY FORMATTED VARIABLE NAMES WITH SYSTEM COLUMN AND COLUMN COMPONENT
## only need to run once
import re
arc_count_df3 = arc_count_df2
arc_count_vars = arc_count_df3.columns
#print(arc_count_vars)
arc_count_vars = list(arc_count_vars)
for i in range(len(arc_count_vars)):
    pattern = r'\{([B-D]{1}[2-4]{1}) Arc Count=\"(.*)\"\}'
    try:
        match = re.findall(pattern, arc_count_vars[i])
        replacement = match[0][0] + ' ' + match[0][1]
        #print(replacement)
        arc_count_vars[i] = replacement
    except:
        print(f"No match for {arc_count_vars[i]}")
arc_count_df3.columns = arc_count_vars
```

```python
# CONVERT "Time" VARIABLE FROM STRING TO DATETIME OBJECT
arc_count_df4 = arc_count_df3
arc_count_df4['Time'] = pd.to_datetime(arc_count_df4['Time'])
arc_count_df4
```

```python
# WRTIE PREPROCESSED ARC COUNT DATAFRAME TO CSV
arc_count_df4.to_csv('data/MEBL3/MEBL3_All_Arc_Counts_Clean', index = False)
```
